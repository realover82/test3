레이어는 Keras에서 신경망의 기본 빌딩 블록입니다 . 레이어는 Tensor-in tensor-out 계산 함수(레이어의 호출 메서드)와 TensorFlow 변수(레이어의 가중치)에 보관된 일부 상태로 구성됩니다.

keras layer api

```python
from keras.layers import *
from keras.models import Model

def build_model():
  x = Input(shape=(64, 64, 3))
  out = x

  out = Flatten()(out)
  out = Dense(512, activation='relu')(out)
  out = Dense(256, activation='relu')(out)
  out = Dense(64, activation='relu')(out)
  out = Dense(1, activation='sigmoid')(out)

  model = Model(x, out)
  return model
```

매개변수의 수는 6,439,809입니다. 모든 은닉 유닛에는 6,439,809개의 입력 가중치와 편향이 있는 연결 가중치가 하나 있기 때문입니다. 이것은 모든 숨겨진 유닛이 6,439,809개의 매개변수를 제공한다는 것을 의미합니다. 

```python
model = build_model()
model.summary()
```

모델 은 훈련 및 추론 기능이 있는 개체로 레이어를 그룹화합니다. ... 모델 과 함께 . compile () , model .fit() 으로 모델을 훈련

```python
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['binary_accuracy'])
history = model.fit(train_gen, epochs=2, steps_per_epoch=len(train_gen))
```

x_test와 y_test를 for문으로 테스트하여 총 몇 장인지 확인한다.(3장확인)

```python
import numpy as np

x_tests, y_tests = [], []
for i in range(len(test_gen)):
  x_test, y_test = next(test_gen)
  x_tests.append(x_test)
  y_tests.append(y_test)

x_test = np.concatenate(x_tests, axis=0)
y_test = np.concatenate(y_tests, axis=0)

print(x_test.shape)
print(y_test.shape)
```
