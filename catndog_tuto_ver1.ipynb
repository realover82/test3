```python
Generator = ImageDataGenerator(rescale=1/255.)

train_gen = Generator.flow_from_directory('catdog/train/train', target_size=(64,64),
                                          batch_size=32,
                                          class_mode='binary')
```

```python
x_batch, y_batch = next(train_gen)
print(x_batch.shape)
print(y_batch.shape)
```

```python
from keras.layers import *
from keras.models import Model

def build_model():
  x = Input(shape=(64, 64, 3))
  out = x

  out = Flatten()(out)
  out = Dense(512, activation='relu')(out)
  out = Dense(256, activation='relu')(out)
  out = Dense(64, activation='relu')(out)
  out = Dense(1, activation='sigmoid')(out)

  model = Model(x, out)
  return model
```

```python
model = build_model()
model.summary()
```

학습

```python
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['binary_accuracy'])
history = model.fit(train_gen, epochs=2, steps_per_epoch=len(train_gen))
```

```python
test_gen = Generator.flow_from_directory ('catdog', target_size=(64, 64),
                                         batch_size=32, classes=['test3'])
```

```python
pred = model.predict(test_gen)
pred.shape
```

```python
x_test, y_test = next(test_gen)
test_gen

for i in range(len(x_test)):
   plt.imshow(x_test[i])
   plt.show()
```

```python
pred = model.predict(x_test[:3])

for i in range(3):
   plt.imshow(x_test[i])
   plt.title(pred[i])
   plt.show()
```

```python
losses = history.history['loss']
accs = history.history['binary_accuracy']

plt.plot(losses, label='loss')
plt.plot(accs, label='acc')
plt.legend()
plt.show()
```
